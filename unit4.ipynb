```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=np.arange(10)
plt.plot(data)
```

```python
fig=plt.figure()
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax4=fig.add_subplot(2,2,4)
plt.plot(np.random.randn(30).cumsum(),'y--',linestyle='-',marker='*')
ax1.hist(np.random.randn(40),bins=20,color='g',alpha=0.3)
ax2.scatter(np.arange(30),np.arange(30),color='orange')
```

```python
fig,axes=plt.subplots(2,3,sharex=True,sharey=True)

axes
plt.subplots_adjust(wspace=0.5,hspace=0.5)

```

```python
data=np.random.randn(30).cumsum()
plt.plot(data,'k--',label='default')
plt.plot(data,'k-',drawstyle='steps-post',label='setp')
plt.legend(loc='best')
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum())
ticks=ax.set_xticks([0,250,500,750,1000])
labels=ax.set_xticklabels(['one','two','three','four','five'],rotation=30,fontsize='small')
prps={
    'title':'my first matplotlib plot',
    'ylabel':'number',
    'xlabel':'stages'
    
    
}
ax.set(**prps)
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
rect=plt.Rectangle((0.2,0.75),0.4,0.15,color='k',alpha=0.3)
circ=plt.Circle((0.7,0.2),0.15,color='b',alpha=1)
pgon=plt.Polygon([[0.15,0.15],[0.35,0.4],[0.2,0.6]],color='g',alpha=0.8)
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```

```python
import pandas as pd
```

```python
s=pd.Series(np.random.randn(10).cumsum(),index=np.arange(0,100,10))
s.plot(xlabel='x ',ylabel='y')
```

```python
df=pd.DataFrame(np.random.randn(10,4).cumsum(0),
                columns=['a','b','c','d'],
                index=np.arange(0,100,10))
df.plot()
```

```python
#barplots
fig,axes=plt.subplots(2,1)
data=pd.Series(np.random.rand(16),index=list('abcdefghijklmnop'))
data.plot.bar(ax=axes[0],color='k',alpha=0.7)
data.plot.barh(ax=axes[1],color='b',alpha=0.5)
```

```python
tips=pd.read_csv('tips.csv')
```

```python
party_counts=pd.crosstab(tips['day'],tips['size'])
party_counts
```

```python
party_counts=party_counts.loc[:,2:5]
party_pcts=party_counts.div(party_counts.sum(1),axis=0)
party_pcts
```

```python
party_pcts.plot.bar()
```

```python
import seaborn as sns
tips['tips_pct']=tips['tip']/(tips['total_bill']-tips['tip'])
tips.head()
```

```python
sns.barplot(x='tips_pct',y='day',data=tips,orient='h')
```

```python
sns.barplot(x='tips_pct',y='day',hue='time',data=tips,orient='h')
sns.set(style='whitegrid')
```
